## 1.

```python
import re
import nltk
from nltk import edit_distance
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.corpus import words

from nltk.stem.snowball import SnowballStemmer
```

```python
nltk.download('stopwords')
```

```python
name = "Olha"
nick = "alloha"
```

```python
edit_distance(name, nick)
```

a) The edit distance between my nickname and my given name is 3.
Seems about right: my given name and nickname share some letters.

```python
print(len(name))
print(len(nick))
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()*100
        
print("Name and nickname are {:.2f}% similar.".format(similar(name, nick)))
```

b) 60% string match between my nickname and my given name.

## 2. 

```python
sentence = "In the beginning God created the heaven and the earth. And the earth was without form, and void; and darkness was upon the face of the deep. And the Spirit of God moved upon the face of the waters. And God said, Let there be light: and there was light."
```

Those sentences were taken from the book of Genesis, found in the Christian bible. I know an individual who has read that book.

```python
tokens = [i for i in re.split("[^a-zA-z]", sentence) if not re.search("[^a-zA-z]", i) and len(i) > 0]

sentence_nostop = [word for word in word_tokenize(sentence) if word.lower() not in stopwords.words("english")]
```

```python
" ".join(sentence_nostop)
```

My friend was able to guess the book from hearing the list of words from 1st try. The first 5 words have made a good hint what a book was used in this test:  "beginning God created heaven earth".

## 3.

```python
all_words = set(words.words())
```

```python
all_words
```

```python
print(tokens)
```

```python
Snowball = SnowballStemmer("english")
s_stems = [Snowball.stem(i) for i in tokens]

valid_counts = 0

for i in s_stems:
    if i in all_words:
        valid_counts += 1
    
frmt = "{:.2f}% of the outputted stems are valid morphological roots of the corresponding words using a Snowball Stemmer."

print(frmt.format(valid_counts / len(s_stems)*100))
```
