### II. Lets create model.

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import linear_model, datasets
from scipy.stats import uniform
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")

# adapt this code below to run your analysis
 
# Recommend to be done before live class 2
# 1. Write a function to take a list or dictionary of clfs and hypers ie use logistic regression, each with 3 different sets of hyper parrameters for each
 
# Recommend to be done before live class 3
# 2. expand to include larger number of classifiers and hyperparmater settings
# 3. find some simple data
# 4. generate matplotlib plots that will assist in identifying the optimal clf and parampters settings
 
# Recommend to be done before live class 4
# 5. Please set up your code to be run and save the results to the directory that its executed from
# 6. Investigate grid search function

def ModelSearch(X,Y):
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=.3,random_state=100, shuffle=True)

    grid1 = [
            {'max_depth': [8, 10]},
            {'max_features': ['auto', 'sqrt', 'log2']},
            {'min_samples_leaf': [2, 4, 12]},
            {'min_samples_split': [2, 6, 12]}
    ]
    rfr = RandomForestClassifier(criterion='entropy',n_estimators=100)
    gs1 = GridSearchCV(rfr, param_grid = grid1)  
    gs1.fit(X_train, y_train)
    y_test_predict = gs1.predict(X_test)
    AUC1 = roc_auc_score(y_test, y_test_predict)
    Acc1 = accuracy_score(y_test, y_test_predict)
    M1 = confusion_matrix(y_test, y_test_predict)
    probs = gs1.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
    roc_auc = metrics.auc(fpr, tpr)
    plt.title('Receiver Operating Characteristic (ROC) curve for Random Forest Classifier')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % AUC1)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    grid2 = [
            {'penalty': ['l1', 'l2'],
            'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
    ]
    rfr = LogisticRegression()
    gs2 = GridSearchCV(rfr, param_grid = grid2)  
    gs2.fit(X_train, y_train)
    y_test_predict = gs2.predict(X_test)
    AUC2 = roc_auc_score(y_test, y_test_predict)
    Acc2 = accuracy_score(y_test, y_test_predict)
    M2 = confusion_matrix(y_test, y_test_predict)
    probs = gs2.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
    roc_auc = metrics.auc(fpr, tpr)
    plt.title('Receiver Operating Characteristic (ROC) curve for Logistic Regression')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % AUC2)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    grid3 = [
            {'n_neighbors': [2, 5, 8, 10],
            'algorithm': ['auto', 'ball_tree', 'brute'],
            'p': [1, 2]}
    ]
    rfr = KNeighborsClassifier()
    gs3 = GridSearchCV(rfr, param_grid = grid3)  
    gs3.fit(X_train, y_train)
    y_test_predict = gs3.predict(X_test)
    AUC3 = roc_auc_score(y_test, y_test_predict)
    Acc3 = accuracy_score(y_test, y_test_predict)
    M3 = confusion_matrix(y_test, y_test_predict)
    probs = gs3.predict_proba(X_test)
    preds = probs[:,1]
    fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
    roc_auc = metrics.auc(fpr, tpr)
    plt.title('Receiver Operating Characteristic (ROC) curve for KNeighbors Classifier')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.3f' % AUC3)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    
    AUC = max(AUC1,AUC2,AUC3)
    if AUC==AUC1:
        technique = "Random Forest Classifier"
        param = gs1.best_params_
    if AUC==AUC2:
        technique = "Logistic Regression" 
        param = gs2.best_params_
    if AUC==AUC3:
        technique = "KNeighbors Classifier" 
        param = gs3.best_params_
        
    print("RandomForestClassifier AUC=",round(AUC1,3) ,"Accuracy=",round(Acc1,3), "Confusion matrix:")
    print(M1)
    print("------------------------------------------------------------------------------------------")
    print("LogisticRegression AUC=",round(AUC2,3) ,"Accuracy=",round(Acc2,3), "Confusion matrix:")
    print(M2)
    print("------------------------------------------------------------------------------------------")
    print("KNeighbors Classifier AUC=",round(AUC3,3) ,"Accuracy=",round(Acc3,3), "Confusion matrix:")
    print(M3)
    print("------------------------------------------------------------------------------------------")
    print("The best technique is",technique ,'with parameters:', param)
    # Load data


df_embed_data = df_embed.drop(columns=['Category'])
X = df_embed_data
y = df_embed.Category
ModelSearch(X,y)
```
